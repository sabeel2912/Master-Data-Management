```python
split_terms = [
    'pvt ltd', ' p ltd', '(p) ltd', 'pvtltd', ' p limited', 'private limited',
    'ltd', 'pltd', 'pvt limited', 'limited', 'private', 'pvt','l td', 'llmited', 'plimited'
]

# Function to split the org_name
def split_org_name(org_name):
    org_name_lower = org_name.lower()
    for term in split_terms:
        term_lower = term.lower()
        split_point = org_name_lower.find(term_lower)
        if split_point != -1:
            part1 = org_name[:split_point].strip()
            part2 = org_name[split_point:].strip()
            return part1, part2
    return org_name, ""

# Apply the function to the DataFrame and update the columns
df['org_name'], df['org_2'] = zip(*df['org_name'].apply(split_org_name))

df
```

```python
import pandas as pd
from rapidfuzz.distance import JaroWinkler


df5 = spmo

# Function to find the more informative name
def more_informative(name1, name2):
    words1, words2 = set(name1.split()), set(name2.split())
    if words1 >= words2:
        return name1
    elif words2 >= words1:
        return name2
    else:
        # If neither name has more information, choose alphabetically earlier one
        return min(name1, name2)

# Function to compute similarities and replace names within each PAN group using Jaro-Winkler similarity
def replace_similar_names(df5, threshold=0.80):
    for pan, group in df5.groupby('pan'):
        org_names = group['org_name'].tolist()
        replace_map = {}
        for i in range(len(org_names)):
            for j in range(i + 1, len(org_names)):
                score = JaroWinkler.similarity(org_names[i], org_names[j])
                if score >= threshold:
                    more_info_name = more_informative(org_names[i], org_names[j])
                    replace_map[org_names[i]] = more_info_name
                    replace_map[org_names[j]] = more_info_name

        # Replace org_name values in the DataFrame
        df5.loc[df5['pan'] == pan, 'org_name'] = df5.loc[df5['pan'] == pan, 'org_name'].replace(replace_map)
    return df5

# Replace similar org_name values in the DataFrame using Jaro-Winkler similarity within each PAN group
df5 = replace_similar_names(df5)
df5
```
